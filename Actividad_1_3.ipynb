```python
import pandas as pd
```

```python
df = pd.read_excel('Inmuebles.xlsx', header = 1)
```

```python
filtro1 = df.drop('Referencia', axis = 1)
filtro1
```

```python
filtro2 = filtro1[(filtro1['Fecha_Alta'] == '2007-04-29') | (filtro1['Fecha_Alta'] == '2004-01-02')]
filtro2
```

```python
filtro1['Tipo'].unique()
filtro3 = filtro1[~(filtro1['Tipo'] == 'Oficina') & ~(filtro1['Tipo'] == 'Suelo') & ~(filtro1['Tipo'] == 'Casa')]
filtro3
```

```python
filtro4 = filtro1[(filtro1['Provincia'] == 'Girona') & (filtro1['Superficie'] > 100)]
filtro4
```

```python
filtro5 = filtro1[(filtro1['Precio Venta'] > 1000000) | (filtro1['Vendedor'] == 'Carmen') | (filtro1['Vendedor'] == 'Joaquin')]
filtro5
```

```python
filtro6 = filtro1[~(filtro1['Provincia'] == 'Barcelona')]
filtro6
```

```python
filtro7 = filtro1[~(filtro1['Operación'] == 'Alquiler') & ~(filtro1['Tipo'] == 'Oficina')]
filtro7
```

```python
from google.colab import files
filtro1.to_csv('Entregable1_3.csv')
filtro2.to_csv('Entregable2_3.csv')
filtro3.to_csv('Entregable3_3.csv')
filtro4.to_csv('Entregable4_3.csv')
filtro5.to_csv('Entregable5_3.csv')
filtro6.to_csv('Entregable6_3.csv')
filtro7.to_csv('Entregable7_3.csv')

files.download('Entregable1_3.csv')
files.download('Entregable2_3.csv')
files.download('Entregable3_3.csv')
files.download('Entregable4_3.csv')
files.download('Entregable5_3.csv')
files.download('Entregable6_3.csv')
files.download('Entregable7_3.csv')
```
